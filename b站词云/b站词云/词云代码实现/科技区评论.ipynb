```python
from wordcloud import WordCloud
import numpy as np
import jieba
from snownlp import SnowNLP
import pandas as pd
from PIL import Image
import os
import jieba.analyse as ana
import nltk
```

```python
data=pd.read_excel('评论区/科技区用户信息.xlsx',header=None,usecols = [6],squeeze  = False)
data
```

```python
text=data.values.tolist()
text=" ".join('%s' %a for a in text)
text=jieba.lcut(text)
```

```python
fdist=nltk.FreqDist(text)
fdist.most_common(500)
```

```python
ana.set_stop_words("停用词.txt")
text=" ".join(text)
text1=ana.extract_tags(text,topK=500)
text1=" ".join(text1)
text1
```
