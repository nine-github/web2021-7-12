```python
from wordcloud import WordCloud
import numpy as np
import jieba
from snownlp import SnowNLP
import pandas as pd
from PIL import Image
import os
import jieba.analyse as ana
import nltk
```

```python
path="影视区弹幕"
filenames=os.listdir(path)
result="影视区弹幕.txt"
f= open(result, 'w+', encoding="utf-8")
```

```python
for filename in filenames:
    filepath = path + '/'
    filepath = filepath + filename
    for line in open(filepath, encoding="utf-8"):
        f.writelines(line)
        f.write('\n')
f.close()
```

```python
file=open(result,"r",encoding='utf-8')
text=file.read()
text=" ".join(jieba.lcut(text))
```

```python
fdist=nltk.FreqDist(text)
fdist.most_common(500)
```

```python
ana.set_stop_words("停用词.txt")
text1=ana.extract_tags(text,topK=1000)
text1=" ".join(text1)
text1
```

```python
mask=np.array(Image.open("电1.jpg"))
mask
```

```python
ys=WordCloud(background_color="white",font_path="Font/msyh.ttc",max_words=500,width=250,height=250,max_font_size=70,mask=mask,colormap="RdPu")
```

```python
ys=ys.generate(text1)
```

```python
ys.to_image()
```

```python
ys.recolor()
```

```python
ys.to_file("影视区弹幕.png")
```
